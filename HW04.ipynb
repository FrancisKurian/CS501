```
# Load library
import numpy as np
import string,sys

colgrid =string.ascii_uppercase[:15] #generate a string of A to O
gridline = [] #define an empty list to hold grid values A1:O15-15X15 matrix

for i in range(15):  #15 rows  A to O
 for n in range(15): #15 columns 1 to 15   
      gridline.append(colgrid[i]+str(n+1)) #creating a long list A1 to O15

gridline_np = np.array( gridline ) # convert the list to a NumPy array for flexibility 
gridline_np = gridline_np.reshape(15,15) # Reshape the NumPy array to 15X15 grid


def word_fits(word,position,across): #define a function with user inputs as parameters 

  if position.upper() in gridline_np : #checking the specified postion is exisiting in the grid
    
    pos=position.upper()
    i, j = np.where(gridline_np==pos) # get the index of user provided position in the grid

    if  across is True  and len(word) <= 15-(i[0]+1) : return True # check the word fits down the grid
    elif across is True  : return False
    elif across is False  and   len(word) <= 15-(j[0]+1) : return True #check the word fits across the grid
    elif across is False : return False
    else:  return False
    #else:  print("Not a valid grid direction(choose:Across =true or false)")

  else:  return False
  #else:  "outside 15X15 grid definition")
```

```
#Test Function to evaluate the word_fits funtion and provide appropriate message

def test(word,position,across):
  across_down='across' if across else 'down'
  mod='not' if not word_fits(word,position,across) else ''
  return '"{}" {} at {} does {} fit'.format(word,across_down,position,mod)
```

```
test('CHAPMAN','B13',across=True)
```

```
test('DATASCIENCE','D6',across=False)
```

```
test('OUTERBOUND15X15','Z10',across=False)
```

```

def pyramid_AV(n,s,h):
  import math
  a=1/2*s*(1/math.tan(math.pi/n))
  l=math.sqrt((h**2)+(a**2))
  base_area=1/2*n*s*a #A
  surface_area=base_area + (1/2*n*s*l)

  volume=1/3*base_area*h
  return volume, surface_area
```

```
n=5
s=36.5
h=12

volume, area =pyramid_AV(n,s,h)
print("volume=",volume)
print("Area=",area)
```
