```
#5 Create a Matrix using NumPy
import numpy as np
A= np.array([[3,7,-4,12],[-5,9,10,2],[6,13,8,11],[15,5,4,1]])
print("Matrix Created using Array A:\n",A)
```

```
#5a: Vector v 

v=A[:, 1].reshape(4,1) #vector 2nd column of array A

#5b: Vector w
w=A[1]
print("5a.Vector using 2nd Column of A v=\n", v)
print("5b.Vector using 2nd row of A w=\n",w)
```

```
#6a 4x3 array b from Matrix A with all elements in 2-4 cloumns
B=A[:,1:]
print("6a.4x3 array b from Matrix A with all elements in 2-4 cloumnsB:\n",B)
#6b 3x4 array C from Matrix A with all elements in 2-4 rows
C=A[1:,:]
print("6b.3x4 array C from Matrix A with all elements in 2-4 rows C:\n",C)
D=A[:2,1:]
print("6c.2x3 array D first 2 rows and last 3 columns from Matrix A D:\n",D)
```

```
#7 Length and abolute value of the given vectors
x= np.array([2,4,7])
y= np.array([2,-4,7])
z= np.array([5+3j,-3+4j,2-7j])

print("7a.Length of vector x:",len(x))
print("7b.Length of vector y:",len(y))
print("7c.Length of vector z:",len(z))

print("7a.Absolute of vector x:\n",abs(x))
print("7b.Absolute of vector y:\n",abs(y))
print("7c.Absolute of vector z:\n",abs(z))
```

```
#10 Create an array A and Array B as ln(A)
A= np.array([[1,4,2],[2,4,100],[7,9,7],[3,np.pi,42]]).reshape(4,3)
print("Matrix A:\n",A)
B=np.log(A)
print("Matrix B=ln(A):\n", B)
print ("10a.Second row of B:\n",B[1,:])

print("10b.Sum of second row of B:\n",np.sum(B[1,:]))

print("10c.Second Column of B: \n",B[:,1].reshape(4,1))
print("First row of A: \n",A[:,0])

C=np.multiply(B[:,1],A[:,0]).reshape(4,1)
print ("10d.Element by elemnt Multiply Second Column of B and First Column of A:\n",C)
print("10e.Element by element division 1st row of A 3rd column B:\n", np.divide(A[0,:],B[:3,2]))
```
