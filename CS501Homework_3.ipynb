```
# create an empty list to populate numbers that meet the criteria

quotients=[]

#use a nested loop 'if' to evaluate the quotients and then populate the list

for i in range(2000,3201):
  if i%7==0 and i%5!=0 :
    quotients.append(i)

print("Numbers divisible by 7 but not a multiple of 5 in 2000-3200 range are:")
print(quotients)
```

```
# Program to find factorial of a given number 
import math 

#take input from the user
numbr = int(input("Enter a number: "))

if numbr >= 0: #check its a non-negative number
  fact = math.factorial(numbr)
  print("The factorial of %d  = %d" %(numbr, fact))
else:
   print("Invalid number.Positive integers expected")
```
